# SENTIMENT SCORING MODEL

## Loading Libraries

```python
import nltk
from nltk.corpus import stopwords
from nltk import word_tokenize
from nltk import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import LancasterStemmer, WordNetLemmatizer

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"


import numpy as np
import pandas as pd

from nltk.stem.porter import PorterStemmer

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

## Reading Dataset

```python
#Reading the data
df = pd.read_excel('Earnings Report.xlsx',dtype=str)
df.head()
```

## Text Preprocessing

```python
#Remove all leading and trailing whitespace at the start or end of a string.
df['Earnings Report'].replace({ r'\A\s+|\s+\Z': '', '\n' : ' '}, regex=True, inplace=True)

#Replace numeric text (0-9) with space
df['Earnings Report'] = df['Earnings Report'].str.replace('\d+', '')

#Replace non-word, and non-space characters with space
df["Earnings Report"] = df['Earnings Report'].str.replace('[^\w\s]','')

#Replace trailing white spaces with only 1-character length white space
df["Earnings Report"] = df['Earnings Report'].str.replace('\s+',' ',regex = True)
```

```python
#Convert to string
df['Earnings Report'] = df['Earnings Report'].astype(str)

#Transform all text to lower case. 
df['Earnings Report'] = df['Earnings Report'].apply(lambda x: " ".join(x.lower() for x in x.split()))

#Remove english stopwords from text string by applying NLTK stopwords
stop = stopwords.words('english')
df['Earnings Report'] = df['Earnings Report'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))

#Apply NLTK PorterStemmer() function to resolve words back to their root word format/context
st = PorterStemmer()
df['Earnings Report'] = df['Earnings Report'].apply(lambda x: " ".join([st.stem(word) for word in x.split()]))
```

## LOADING VADER - SENTIMENT SCORING MODEL

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer= SentimentIntensityAnalyzer()
```

```python
#Dertermining Sentiment Scores for the earnings report news
df['compound'] = [analyzer.polarity_scores(x)['compound'] for x in df['Earnings Report']]
df['neg'] = [analyzer.polarity_scores(x)['neg'] for x in df['Earnings Report']]
df['neu'] = [analyzer.polarity_scores(x)['neu'] for x in df['Earnings Report']]
df['pos'] = [analyzer.polarity_scores(x)['pos'] for x in df['Earnings Report']]
df
```

## Setting a Threshold for Sentiment Score

```python
def f(row):
    if row['compound'] > 0.80:
        val = "Positive"
    else:
        val = "Negative"
    return val

df['Sentiment'] = df.apply(f, axis=1)
df
```
